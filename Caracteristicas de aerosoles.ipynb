```python
import matplotlib.pyplot as plt
import pandas as pd

# Supongamos que tu dataframe merge_sat es similar a este ejemplo
data = {
    'SDA_mean': [0.7, 0.75, 0.8, 0.85, 0.9, 0.95, 1.0, 0.82, 0.78, 0.86, 0.92],
    'fmf_mean': [0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.58, 0.62, 0.67, 0.73],
    'season': ['Spring', 'Summer', 'Fall', 'Winter', 'Spring', 'Summer', 'Fall', 'Winter', 'Spring', 'Summer', 'Fall']
}
merge_sat = pd.DataFrame(data)

# Colores y formas personalizados
colors = {'Spring': 'green', 'Summer': 'red', 'Fall': 'blue', 'Winter': 'black'}
shapes = {'Spring': 's', 'Summer': '^', 'Fall': 'v', 'Winter': 'X'}

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Graficar los puntos
for season in merge_sat['season'].unique():
    season_data = merge_sat[merge_sat['season'] == season]
    ax.scatter(season_data['SDA_mean'], season_data['fmf_mean'], 
               color=colors[season], marker=shapes[season], label=season)

# Líneas verticales y horizontales
ax.axvline(x=0.95, linestyle='dashed', color='black')
ax.axhline(y=0.4, linestyle='dashed', color='black')
ax.axhline(y=0.6, linestyle='dashed', color='black')
ax.plot([0.85, 0.85], [0.6, 1], linestyle='dashed', color='black')
ax.plot([0.9, 0.9], [0.6, 1], linestyle='dashed', color='black')

# Añadir etiquetas de texto
ax.text(0.75, 0.62, 'FH-A', size=10)
ax.text(0.87, 0.62, 'FM-A', size=10)
ax.text(0.93, 0.62, 'FS-A', size=10)
ax.text(0.97, 0.62, 'FN-A', size=10)
ax.text(0.75, 0.42, 'M-A', size=10)
ax.text(0.97, 0.42, 'MN-A', size=10)

# Personalizar los ejes
ax.set_xlim(0.7, 1.0)
ax.set_ylim(0.4, 1.0)
ax.set_xticks([0.7, 0.75, 0.8, 0.85, 0.9, 0.95, 1.0])
ax.set_yticks([0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1.0])

# Añadir títulos y etiquetas
ax.set_title('SP')
ax.set_xlabel('SDA[440]')
ax.set_ylabel('FMF')

# Añadir leyenda y quitar la leyenda de la figura
ax.legend(title='Season')
ax.legend().remove()

# Aplicar estilo minimalista
plt.style.use('seaborn-whitegrid')

# Mostrar el gráfico
plt.show()
```

```python
def caracteristicas_aerosoles_plot (path, nrows,ncols, resolucion_espaciotemporal):

    i=0 
    j=0 
    colors = {'Spring': 'green', 'Summer': 'red', 'Fall': 'blue', 'Winter': 'black'}
    shapes = {'Spring': 's', 'Summer': '^', 'Fall': 'v', 'Winter': 'X'}
    
    fig, ax = plt.subplots(ncols=ncols,nrows=nrows,  figsize=(16,8),dpi=500)
    fig.suptitle(resolucion_espaciotemporal, fontsize=16, fontweight='bold')
    #The folder where the files of each station are located is browsed.
    for p,files in enumerate(os.listdir(path)):
        print(p)
        data = pd.read_csv(os.path.join(path,files))
        station_name = files [3:5]
#        # Station Name
        ########################                VARIABLES                ########################
        for season in data['season'].unique():
            season_data = data[data['season'] == season]
            ax[i][j].scatter(season_data['SDA_mean'], season_data['fmf_mean'], 
                       color=colors[season], facecolors='none',marker=shapes[season], label=season)
            
                # Líneas verticales y horizontales
        ax[i][j].axvline(x=0.95, linestyle='dashed', color='black', linewidth=0.8)
        ax[i][j].axhline(y=0.4, linestyle='dashed', color='black', linewidth=0.8)
        ax[i][j].axhline(y=0.6, linestyle='dashed', color='black', linewidth=0.8)
        ax[i][j].plot([0.85, 0.85], [0.6, 1], linestyle='dashed', color='black', linewidth=0.8)
        ax[i][j].plot([0.9, 0.9], [0.6, 1], linestyle='dashed', color='black', linewidth=0.8)

        # Añadir etiquetas de texto
        ax[i][j].text(0.75, 0.62,'FH-A',fontdict={'family': 'serif', 
            'color' : 'Black',
            'weight': 'bold',
            'size': 9})# 'FH-A', size=10)
        ax[i][j].text(0.86, 0.62, 'FM-A',fontdict={'family': 'serif', 
            'color' : 'Black',
            'weight': 'bold',
            'size': 9})
        ax[i][j].text(0.91, 0.62, 'FS-A',fontdict={'family': 'serif', 
            'color' : 'Black',
            'weight': 'bold',
            'size': 9})
        ax[i][j].text(0.96, 0.62, 'FN-A',fontdict={'family': 'serif', 
            'color' : 'Black',
            'weight': 'bold',
            'size': 9})
        ax[i][j].text(0.75, 0.42, 'M-A',fontdict={'family': 'serif', 
            'color' : 'Black',
            'weight': 'bold',
            'size': 9})
        ax[i][j].text(0.96, 0.42, 'MN-A',fontdict={'family': 'serif', 
            'color' : 'Black',
            'weight': 'bold',
            'size': 9})

        # Personalizar los ejes
        ax[i][j].set_xlim(0.7, 1.0)
        ax[i][j].set_ylim(0.4, 1.0)
        ax[i][j].set_xticks([0.7, 0.75, 0.8, 0.85, 0.9, 0.95, 1.0])
        ax[i][j].set_yticks([0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1.0])

        # Añadir títulos y etiquetas
        ax[i][j].set_title('SP')
        ax[i][j].set_xlabel('SSA[440]')
        ax[i][j].set_ylabel('FMF')

        # Añadir leyenda y quitar la leyenda de la figura
        ax[i][j].legend(title='Season',frameon = True, markerfirst = False, fontsize = 12,loc='upper left')
        ax[i][j].legend().remove()

        # Aplicar estilo minimalista
        plt.style.use('seaborn-whitegrid')

        ax[i][j].set_title(station_name,fontsize=13,fontdict={'family': 'serif', 
            'color' : 'Black',
            'weight': 'bold',
            'size': 16})

        plt.subplots_adjust(left=0.125,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.2, 
                    hspace=0.4)
       
        
         
        # Location of data in each subplot
        if j == ncols-1:
            i = i+1
            j = 0
        else:
            j = j+1
   # Save the plot to the current path
    plot = "no"
    if plot == 'save':
        return(plt.savefig("linear_regression_plot.jpg", dpi=500))
    else:
        return(plt.show())
```
